```python
#changed the name of the Title column to 'name' so it matches the other df

listings_df_dropped.rename(columns={"Title": "name", "ID": "listing_id"}, inplace=True)

listings_df_dropped
```

```python
# drop useless columns from Reviews df

reviews_df0.drop(['query', 'google_id', 'rating', 'place_id', 'location_link', 'reviews_per_score', 'reviews', 'review_id', 'author_link',
       'owner_answer', 'owner_answer_timestamp',
       'owner_answer_timestamp_datetime_utc', 'review_likes',
       'reviews_id'], axis=1, inplace=True)
reviews_df = reviews_df0
```

```python
reviews_df
```

```python
# convert the values in col1 to integers
reviews_df['review_rating'] = reviews_df['review_rating'].astype(str)

# display the dataframe
reviews_df['review_rating']
```

```python
reviews_df['rating'] = reviews_df['review_rating'].apply(lambda x: x[0])

reviews_df['rating']
```

```python
# vllokup to add listing id to reviews dataframe

reviews_df1 = pd.merge(reviews_df, 
                      listings_df_dropped, 
                      on ='name', 
                      how ='inner')
reviews_df1
```

```python
# Import Users (review writers)

reviews_df1[['first_name','last_name']] = reviews_df1['author_title'].loc[reviews_df1['author_title'].str.split().str.len() == 2].str.split(expand=True)
reviews_df1
```

```python
reviews_df1['last initial'] = reviews_df1['last_name'].astype(str).apply(lambda x: x[0].upper() + '.')

reviews_df1
```

```python
# Define a custom function to extract the last 6 digits
def extract_last_6_digits(x):
    return str(x)[-6:]

# Apply the custom function to the 'author_id' column and store the result in a new column 'last_6_digits'
reviews_df1['author_id_last_6'] = reviews_df1['author_id'].apply(extract_last_6_digits)

# Print the updated dataframe
reviews_df1
```

```python
reviews_df1.columns
```

```python
# define a custom function to concatenate values
def author_username(row):
    return str(row['first_name']) + str(row['author_id_last_6'])

# apply the custom function to each row of the dataframe
reviews_df1['author_username'] = reviews_df1.apply(author_username, axis=1)

# display the dataframe
reviews_df1
```

```python
# define a custom function to concatenate values
def author_email(row):
    return str(row['author_username']) + '@gmail.com'

# apply the custom function to each row of the dataframe
reviews_df1['author_email'] = reviews_df1.apply(author_email, axis=1)

# display the dataframe
reviews_df1
```

```python
lp_options_1 = 'a:9:{s:7:"gallery";s:0:"";s:6:"rating";s:1:"'
lp_options_2 = '";s:10:"listing_id";s:6:"'
lp_options_3 = '";s:12:"review_reply";s:0:"";s:18:"review_Interesting";s:0:"";s:10:"review_Lol";s:0:"";s:11:"review_Love";s:0:"";s:18:"review_reported_by";s:0:"";s:15:"review_reported";s:0:"";}'
```

```python
# define a custom function to concatenate values
def lp_options(row):
    return lp_options_1 + str(row['rating']) + lp_options_2 + str(row['listing_id']) + lp_options_3

# apply the custom function to each row of the dataframe
reviews_df1['lp_options_full'] = reviews_df1.apply(lp_options, axis=1)

# display the dataframe
reviews_df1['lp_options_full'] = reviews_df1['lp_options_full'].astype(str)

reviews_df1['lp_options_full']
```
